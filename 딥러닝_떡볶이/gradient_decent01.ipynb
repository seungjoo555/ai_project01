```python
import pandas as pd

df = pd.DataFrame({
    "temperature":[10, 7, 4, 0],
    "sales": [30, 50, 60, 80]
})

display(df)
```

```python
a = -4
b = 80

for i in range(len(df)):
    x = df.loc[i, "temperature"]
    pred = a * x + b
    print(f"X={x}일 때 예측값 = {a} * {x} + {b} = {pred}")
```

```python
a = -4
b = 80

for i in range(len(df)):
    x = df.loc[i, "temperature"]
    y = df.loc[i, "sales"]
    pred = a * x + b
    error = pred -y

    print(f"X={x}, 실제값={y}, 예측값={pred}, 오차={pred}-{y}={error}")
    print("="*30)
```

```python
a = -4
b = 80

square_sum = 0

for i in range(len(df)):
    x = df.loc[i, "temperature"]
    y = df.loc[i, "sales"]
    pred = a * x + b
    error = pred - y
    error_square = error * error
    square_sum = square_sum + error_square

n = len(df)
cost = square_sum / n

print(f"오차 제곱 합: {square_sum}")
print(f"데이터 개수: {n}")
print(f"Cost: {cost:.4f}")
```

```python
def calcculate_cost(a, b):
    n = len(df)
    print(f"n={n}")
    square_sum = 0
    
    for i in range(n):
        x = df.loc[i, "temperature"]
        y = df.loc[i, "sales"]
        print(f"x={x}, y={y}")

        pred = a * x + b
        print(f"pred={pred}")

        error = y - pred
        print(f"error={error}")

        error_square = error * error
        print(f"error_square={error_square}")

        square_sum = square_sum + error_square
        print(f"square_sum={square_sum}")
    
    n = len(df)
    cost = square_sum /n
    return cost

print("calculate_cost(-4, 80) =", calculate_cost(-4, 80))
```

```python
cost1 = calculate_cost(-4, 80)
cost2 = calculate_cost(-4.9, 81)

print(f"Y = -4X + 80 의 Cost: {cost1:.4f}")
print(f"Y = -4.9X + 81 의 Cost: {cost2:.4f}")
```

```python
a_values = [-6, -5, -4, -3, -2]
b = 80

cost_values = []

for a in a_values:
    cost = calculate_cost(a, b)
    cost_values.append(cost)
    print(f"a={a}, b={b}, Cost={cost:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(a_values, cost_values, marker='o')
plt.title("a value vs Cost")
plt.xlabel("a value")
plt.ylabel("Cost")
plt.grid(True)
plt.show()
```
