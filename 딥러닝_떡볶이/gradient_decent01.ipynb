```python
import pandas as pd

df = pd.DataFrame({
    "temperature":[10, 7, 4, 0],
    "sales": [30, 50, 60, 80]
})

display(df)
```

```python
a = -4
b = 80

for i in range(len(df)):
    x = df.loc[i, "temperature"]
    pred = a * x + b
    print(f"X={x}일 때 예측값 = {a} * {x} + {b} = {pred}")
```

```python
a = -4
b = 80

for i in range(len(df)):
    x = df.loc[i, "temperature"]
    y = df.loc[i, "sales"]
    pred = a * x + b
    error = pred -y

    print(f"X={x}, 실제값={y}, 예측값={pred}, 오차={pred}-{y}={error}")
    print("="*30)
```
